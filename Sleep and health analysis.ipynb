```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
df
```

```python
g=sns.PairGrid(df)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
g=sns.PairGrid(df,hue="BMI Category")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
```

```python
g=sns.PairGrid(df,hue="BMI Category",vars=["Stress Level","BMI Category"])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.show()
```
